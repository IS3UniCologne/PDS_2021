# Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds_df = sns.load_dataset("diamonds")
```

```python
diamonds_df.sample(5)
```

```python
diamonds_df.info()
```

```python
diamonds_df.describe()
```

```python
diamonds_df["cut"]
```

```python
diamonds_df["cut"].unique()
```

```python
diamonds_df["cut"].nunique()
```

```python
sns.histplot(diamonds_df["price"], bins=10, kde=True)
plt.show()
```

```python
sns.jointplot(x = "carat", y = "price", data=diamonds_df)
plt.show()
```

```python
sns.pairplot(diamonds_df)
plt.show()
```

```python
sns.barplot(x="cut", y="price", data=diamonds_df, estimator=np.mean)
plt.show()
```

```python
sns.countplot(x="cut", data=diamonds_df)
plt.show()
```

```python
sns.boxplot(x="cut", y="price", data=diamonds_df)
plt.show()
```

```python
sns.violinplot(x="cut", y="price", data=diamonds_df)
plt.show()
```

```python
diamonds_df_ED = diamonds_df.loc[diamonds_df["color"].isin(["E", "D"])]
```

```python
diamonds_df_ED["color"].unique()
```

```python
diamonds_df_ED["color"]
```

```python
diamonds_df_ED.color = diamonds_df_ED.color.cat.remove_unused_categories()
```

```python
diamonds_df_ED["color"]
```

```python
sns.violinplot(x="cut", y="price", data=diamonds_df_ED, hue="color", split=True)
plt.show()
```

```python
diamonds_df.corr()
```

```python
sns.heatmap(diamonds_df.corr())
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="carat", y="price", hue="color", data=diamonds_df_ED, palette="coolwarm")

plt.show()
```
