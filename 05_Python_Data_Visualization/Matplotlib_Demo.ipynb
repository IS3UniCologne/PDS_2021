# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(-2*np.pi,  2*np.pi, 1000)
xx
```

```python
yy = np.sin(xx)
```

```python
plt.plot(xx, yy)
plt.show()
```

```python
plt.plot(xx, yy, 'r')
plt.show()
```

```python
plt.plot(xx, yy, 'r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine Curve")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6), dpi=200)

ax.plot(xx, yy, 'r', label="Sine")

for k in range(1,10):
    if k != 5: # k = 1, 2, 3, 4, 6, 7, 8, 9
        ax.plot(xx, yy*k/5, 'r--', linewidth=0.3*(1-((k/5)-1)**2))

yy2 = np.cos(xx)
ax.plot(xx, yy2, 'b', label="Cosine")

for k in range(1,10):
    if k != 5: # k = 1, 2, 3, 4, 6, 7, 8, 9
        ax.plot(xx, yy2*k/5, 'b--', linewidth=0.3*(1-((k/5)-1)**2))
        
ax.set_xlim([-2*np.pi, 2*np.pi])

ax.set_ylabel("y")
ax.set_xlabel("x")

ax.set_title("Sine and Cosine")

ax.legend()
        
plt.show()
```

```python
x = np.random.randn(100)
y = 2*x + np.random.randn(100)
z = np.abs(x**2+2*y**3+np.random.randn(100))
c = np.abs(z*y+np.random.randn(100))
```

```python
data = np.array([x, y, z, c]).transpose()
```

```python
data.shape
```

```python
df = pd.DataFrame(data=data, columns=["x", "y", "z", "c"])
df.head()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6), dpi=200)

scatter = ax.scatter(x = df["x"], y = df["y"], s = df["z"], c = df["c"])

ax.set_xlabel("x")
ax.set_ylabel("y")

cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label("c")

size_handles, size_labels = scatter.legend_elements(prop="sizes", alpha=0.5)

ax.legend(size_handles, size_labels, loc="lower right", title="z")

plt.show()
```

```python
df = pd.DataFrame(data=data, columns=["x", "y", "z", "c"])
df
```

```python
df[df["z"]>=50]
```

```python
#df = pd.DataFrame(data=data, columns=["x", "y", "z", "c"])
df_list = []

df_list.append(df[df["z"]>=50])
df_list.append(df[df["z"]<50])

markers = ['o', '*']

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6), dpi=200)

for i, df in enumerate(df_list):
    scatter = ax[i].scatter(x=df["x"], y=df["y"], s=df["z"], c=df["c"], marker=markers[i])
    
    ax[i].set_xlabel("X")
    ax[i].set_ylabel("Y")

    scatter.set_clim([0,4000])
    cbar = fig.colorbar(scatter, ax=ax[i])
    cbar.set_label("C")

    size_handles, size_labels = scatter.legend_elements(prop="sizes", alpha=0.5)
    ax[i].legend(size_handles, size_labels, loc="lower right", title="Z")
    
plt.show()
```
